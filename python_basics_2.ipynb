```python
# ==============================
# 1. PRINT & COMMENTS
# ==============================
print("Hello, Python!")   # This is a comment

# ==============================
# 2. VARIABLES & DATA TYPES
# ==============================
name = "Tripti"
age = 20
height = 5.4
is_student = True

print(name, age, height, is_student)

# ==============================
# 3. TYPE CHECKING
# ==============================
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))

# ==============================
# 4. USER INPUT
# ==============================
user_name = input("Enter your name: ")
print("Welcome,", user_name)

# ==============================
# 5. TYPE CASTING
# ==============================
a = int("10")
b = float("5.5")
c = str(100)

print(a, b, c)

# ==============================
# 6. OPERATORS
# ==============================
x = 10
y = 3

print(x + y)   # Addition
print(x - y)   # Subtraction
print(x * y)   # Multiplication
print(x / y)   # Division
print(x % y)   # Modulus
print(x ** y)  # Power
print(x // y)  # Floor division

# ==============================
# 7. IF - ELSE
# ==============================
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Fail")

# ==============================
# 8. LOOPS
# ==============================
# FOR LOOP
for i in range(1, 6):
    print("For Loop:", i)

# WHILE LOOP
count = 1
while count <= 5:
    print("While Loop:", count)
    count += 1

# ==============================
# 9. BREAK & CONTINUE
# ==============================
for i in range(1, 6):
    if i == 3:
        continue
    if i == 5:
        break
    print(i)

# ==============================
# 10. LIST
# ==============================
numbers = [10, 20, 30, 40]
numbers.append(50)
numbers.remove(20)

print(numbers)
print(numbers[0])

# ==============================
# 11. TUPLE
# ==============================
colors = ("red", "green", "blue")
print(colors[1])

# ==============================
# 12. SET
# ==============================
unique_numbers = {1, 2, 3, 3, 4}
print(unique_numbers)

# ==============================
# 13. DICTIONARY
# ==============================
student = {
    "name": "Tripti",
    "age": 20,
    "course": "CSE"
}

print(student)
print(student["name"])

# ==============================
# 14. FUNCTIONS
# ==============================
def add(a, b):
    return a + b

print("Sum:", add(5, 3))

# ==============================
# 15. DEFAULT ARGUMENT
# ==============================
def greet(name="User"):
    print("Hello", name)

greet()
greet("Tripti")

# ==============================
# 16. LAMBDA FUNCTION
# ==============================
square = lambda x: x * x
print("Square:", square(5))

# ==============================
# 17. STRING OPERATIONS
# ==============================
text = "python programming"

print(text.upper())
print(text.lower())
print(text.title())
print(text.split())

# ==============================
# 18. FILE HANDLING
# ==============================
file = open("demo.txt", "w")
file.write("Hello File")
file.close()

file = open("demo.txt", "r")
print(file.read())
file.close()

# ==============================
# 19. EXCEPTION HANDLING
# ==============================
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
finally:
    print("Execution complete")

# ==============================
# 20. IMPORT MODULE
# ==============================
import math

print(math.sqrt(16))
print(math.factorial(5))

# ==============================
# 21. MAIN CHECK
# ==============================
if __name__ == "__main__":
    print("Program executed directly")
```
